```python
# Use a pipeline as a high-level helper
from transformers import pipeline
```

```python
# load model
classifier = pipeline("text-classification", model="mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis")

def getSentiment(s):
  return classifier(s)
```

```python
summarizer = pipeline("summarization")
def getSummary(s):
    return summarizer(s)
```

```python
df = pd.read_csv('recover_points.csv')
```

```python
df.head()
```

```python
df["DropSentiment"] = "None"
df["RecoverySentiment"] = "None"
for index, row in df.iterrows():
    # get the string in the 'Date' column
    drop_day = row['drop_day']
    if drop_day!="None":
      st = getSentiment(drop_day)[0]['label']
      # print(st)
      df.at[index, 'DropSentiment'] = st

    recover_day = row['recover_day']
    if recover_day!="None":
      st = getSentiment(drop_day)[0]['label']
      # print(st)
      df.at[index, 'RecoverySentiment'] = st
```
