```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf

# suppress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#download data for the stock 'INFY' for last 5 years from yahoo finance 
data = yf.download('ACC.NS', '2021-01-01', '2022-01-01')
```

```python
data.head()
```

```python
#calculate returns
data['returns'] = data['Adj Close'].diff()
data.head()
```

```python
col_name = 'Close'
```

```python
#calcualte standard deviation in the std_dev column for last 20 days
data['std_dev'] = data[col_name].rolling(20).std()
# calculate mean in the mean column for last 20 days
data['mean'] = data[col_name].rolling(20).mean()
data.head(30)
```

```python
#make a column called cum_returns which will cumulatively sum the returns
data['cum_returns'] = data['returns'].cumsum()
data.head()
```

```python
# add a column called drop with default value 0
data['drop'] = 0
data.head(30)
```

```python
for i in range(20,len(data)):
    if data[col_name][i] - data['mean'][i-20] <= -2*data['std_dev'][i]:
        data['drop'][i] = 1
data.head(30)
```

```python
# print the drop dates where drop is 1
data[data['drop']==1].index
```

```python
# add column recover with default value 0
data['recover'] = 0
```

```python
# get the index of ith row
data.index[0]
```

```python
# check if atleast 90% of the drop of value is recovered within any of next 5 days
# new data frame with recovery dates and corresponding drop dates

recover_points = pd.DataFrame(columns=['drop_date','recover_date'])

for i in range(len(data)-5):
    if data['drop'][i] == 1:
        drop_val = data['mean'][i] - data[col_name][i]
        for j in range(1,6):
            if data[col_name][i+j] >= data[col_name][i] + 0.9*drop_val:
                recover_points = recover_points.append({'drop_date':data.index[i],'recover_date':data.index[i+j]},ignore_index=True)
                data['recover'][i+j] = 1
                break
data.head(30)
recover_points.head(30)
```

```python
# remove rows with duplicate recover dates
recover_points = recover_points.drop_duplicates(subset='recover_date',keep='first')

# print the length of the recover_points data frame
print(len(recover_points))

# save the data frame to csv file
recover_points.to_csv('recover_points.csv')
```

```python
# print the recover dates where recover is 1
data[data['recover']==1].index
```

```python
# plot the returns
plt.figure(figsize=(100,10))
plt.plot(data.index, data[col_name])



# add red dots at the drop dates where drop is 1
plt.plot(data[data['drop']==1].index, data[data['drop']==1][col_name], 'ro')

# add green dots at the recover dates where recover is 1
plt.plot(data[data['recover']==1].index, data[data['recover']==1][col_name], 'go')

#add labels to axes 
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```
